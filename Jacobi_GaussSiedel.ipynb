```python
import numpy as np

def jacobi(a, b ,x):        
  n = len(a)     
  y = np.zeros(n) # membuat wadah tampungan solusi sementara
  for j in range(0, n):         
    # d adalah variabel untuk menyimpan b[j] 
    d = b[j]

    # menghitung xi(penyelesaian) 
    for i in range(0, n):      
      if(j != i): #diagonal tidak dihitung
        d-=a[j][i] * x[i] 
        
    # update nilai solusi sementara agar tidak menggangu x
    y[j] = d / a[j][j]

  # update nilai solusi ke x
  for j in range (0,n):  
    x[j] = y[j]

  # return solusi
  return x     
                         
a = [[4, -2, 1],[1, -3, 2],[-1, 2, 6]] # Persamaan
b = [1.0, 2.0, 3.0] # Hasil
x = [1.0, 1.0, 1.0] # inisialisasi solusi dengan x = (1,1,1)

print("Persamaan A = ",a)
print("Hasil B = ",b)
print("x = ",x)
print("")

print("Proses Iterasi :")
for i in range(1, 300): # melakukan iterasi
    x = jacobi(a, b, x)
    print(i, x)
print("\nSolusi = ",x)
```

```python
def gausssiedel(a, b ,x):        
  n = len(a)                    
  for j in range(0, n):         
    # d adalah variabel untuk menyimpan b[j] 
    d = b[j]                   

    # menghitung xi(penyelesaian) 
    for i in range(0, n):      
      if(j != i): #diagonal tidak dihitung atau dianggap 0
        d-=a[j][i] * x[i] 
        
    # update nilai solusi      
    x[j] = d / a[j][j]

  # return solusi
  return x     

a = [[4, -2, 1],[1, -3, 2],[-1, 2, 6]] # Persamaan
b = [1.0, 2.0, 3.0] # Hasil
x = [1.0, 1.0, 1.0] # inisialisasi solusi dengan x = (1,1,1)

print("Persamaan A = ",a)
print("Hasil B = ",b)
print("x = ",x)
print("")

print("Proses Iterasi :")
for i in range(1, 40): # melakukan iterasi
    x = gausssiedel(a, b, x)
    print(i, x)
print("\nSolusi = ",x)
```
